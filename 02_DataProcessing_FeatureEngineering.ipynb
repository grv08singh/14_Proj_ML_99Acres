```python
df = (
    pd
    .read_csv("01_flats.csv")    #reading data from csv file
    
    .drop_duplicates()    #removing duplicate rows if present
    
    .dropna(subset=['price','society','areaWithType','bathroom','balcony',
                    'floorNum','agePossession','nearbyLocations','description'])    #removing rows where mandatory data is NA
    
    .apply(lambda col: col.str.strip().str.lower() if col.dtype=='object' else col)    #all data to lower case
    
    .assign(
        bhk = lambda df_: df_['property_name'].str.extract(r"(\d{1})").astype('int8')    #extracting the value of BHK
        
        ,locality = lambda df_: df_['property_name'].str.split('in ').str[1]    #extracting the name of locality
        
        ,society = lambda df_: df_['society'].str.split(r"(\d{1}\.{1}\d*)").str[0]    #extracting the name of society
        
        ,price_cr = lambda df_: df_['price'].str.replace('₹','').apply(
                                lambda x: x.replace('crore','').strip() if 'crore' in x 
                                    else round(float(x.replace('lac','').strip())/100,2) if 'lac' in x 
                                        else np.nan).astype('float')    #price in crore Rupees
        
        ,bathroom = lambda df_: df_['bathroom'].str.extract(r"(\d+)").astype('int8')    #number of bathrooms
        
        ,balcony = lambda df_: df_['balcony'].str.extract(r"(\d*)").replace('','0').astype('int8')    #number of balconies
        
        ,area_sqft = lambda df_: df_['areaWithType'].str.extract(r"(\d*.?\d+)").astype('float')    #intermediate step
        ,area_sqmm = lambda df_: df_['areaWithType'].str.extract(r"(\(\d+\.?\d+\s*.*m{1}|M{1})")    #intermediate step
        
        ,area_sqm = lambda df_: df_['area_sqmm'].str.extract(r"(\d+.?\d+)")
                                .astype('float').fillna(df_['area_sqft']/10.764)    #extracting the area in sq.m.
        
        ,area_type = lambda df_: df_['areaWithType'].str.split(' area').str[0].str.strip()    #type of area
        
        ,addl_room = lambda df_: np.where(df_['additionalRoom'].notna(), 1, 0).astype('int8')    #if additional room is present
        
        ,floor_no = lambda df_: df_['floorNum'].str.extract(r"(\d+)").fillna('0').astype(int)    #floor no. of the flat
        
        ,floor_rise = lambda df_: np.where(df_['floor_no']<4, 'low', 
                                      np.where(df_['floor_no']<8, 'mid', 'high'))    #floor-rise of the flat
        
        ,age_1 = lambda df_: df_['agePossession'].str.replace('within 3 months','0').str.
                                replace('within 6 months','0').str.findall(r"(\d+)").str[0].fillna(-1).astype(int).
                                    apply(lambda x: 2026-x if x>1000 else x)    #age in years
        
        ,age = lambda df_: np.where(df_['age_1']<0, 'under construction',
                                    np.where(df_['age_1']<2, 'new', 
                                        np.where(df_['age_1']<5, 'mid aged', 'old')))    #relative age
        
        ,nearby_loc = lambda df_: df_['nearbyLocations'].apply(lambda x: re.findall(r"'(.*?)'",str(x)))    #list of nearby loc.
    )
    
    .drop(columns=['property_name','link','property_id','bedRoom','price',
                   'areaWithType','area_sqft','area_sqmm','additionalRoom',
                   'floorNum','floor_no','agePossession','age_1','nearbyLocations',
                   'address','description','facing','features'])    #removing unnecessary columns
    
    .reindex(columns=['price_cr','bhk','bathroom','balcony','addl_room','floor_rise','area_sqm',
                      'area_type','age','society','locality','furnishDetails','nearby_loc'])    #rearranging columns
    
    .reset_index(drop=True)    #resetting index
)

df.sample(3)
```

```python
df.to_csv('03_flats_cleaned.csv', index=False)
```
