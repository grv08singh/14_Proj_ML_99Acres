```python
general_landmarks = ['mandir','bank','school','hospital','clinic','atm','nursing','dental',
                    'medical','eye','metro','station','railway','pizza','college','golf',
                    'chowk','lake','mall','highway','hwy','rly','railway','university',
                    'airport','community','petrol','cng','expressway','institute',
                    'park','healthcare','plaza','resort','waterpark','stadium','club',
                    'complex','pharmacy','inn','imt','academy','motors','dhaba','junction',
                    'tower','hilton','trauma','centre','center','dental','domino','moti mahal',
                    'pind baluchi','church','petroleum','reliance','naval','mcdonald','bar',
                    'liquor','wine','beer','whiskey','rum','scotch','cafe','distillery','hotel',
                    'spa','sagar ratna','store','national','birla','hyatt','lemontree','lounge',
                    'bakery','oil','coffee','starbucks','belgian','waffle','petrol','cng','isbt',
                    'sports','hill','pvr','helipad','huda','cinema','bazaar','farm','expy']
```

```python
def landmark_clean(row):
    row = row.replace('[','').replace(']','').replace("'","").replace("\\","").split(",")
    row_ret = []
    for item in row:
        for gen_lmk in general_landmarks:
            if gen_lmk in item:
                row_ret.append(item)
    return row_ret
```

```python
df['landmarks'] = df['nearby_loc'].apply(landmark_clean)
df.drop(columns=['nearby_loc'], inplace=True)
```

```python
for row in df['furnishDetails']:
    row = str(row).replace('[','').replace(']','').replace("'","").replace('no','0').strip().split(",")
    for item in row:
        if item != 'nan':
            name_obj = re.findall(r"([a-zA-Z/s]+)", item)
            if name_obj is not None:
                name = " ".join(name_obj)
                
            count_obj = re.search(r"(\d+)", item)
            if count_obj is not None:
                count = count_obj.group()
                df[name]=count

df.drop(columns=['furnishDetails'], inplace=True)
```
